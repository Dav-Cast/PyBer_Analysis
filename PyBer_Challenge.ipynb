### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
#%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt

# File to Load (Remember to change these)
city_data_to_load = "./Resources/city_data.csv"
ride_data_to_load = "./Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.tail()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_type = pyber_data_df.groupby("type")["ride_id"].count()
total_rides_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_type = city_data_df.groupby("type")["driver_count"].sum()
total_drivers_type
```

```python
#  3. Get the total amount of fares for each city type
total_fares_type = pyber_data_df.groupby("type")["fare"].sum()
total_fares_type
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_ride = total_fares_type / total_rides_type
avg_fare_ride
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_driver = total_fares_type / total_drivers_type
avg_fare_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
    "Total Rides" : total_rides_type,
    "Total Drivers" : total_drivers_type,
    "Total Fares" : total_fares_type,
    "Average Fare per Ride" : avg_fare_ride,
    "Average Fare per Driver" : avg_fare_driver
})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].apply(lambda x: "${:,.2f}".format((x)))
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].apply(lambda x: "${:,.2f}".format((x)))
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].apply(lambda x: "${:,.2f}".format((x)))
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].apply(lambda x: "{:,}".format((x)))
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].apply(lambda x: "{:,}".format((x)))
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
fares_date_df = pyber_data_df.groupby(["type","date"])[["fare"]].sum()
#fares_date_df = pd.DataFrame(fares_date_df)
fares_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
fares_date_df = fares_date_df.reset_index()
fares_date_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_fares_by_date = fares_date_df.pivot(index="date", columns="type", values="fare")
pivot_fares_by_date
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_april_df = pivot_fares_by_date.loc['2019-01-01':'2019-04-29']
jan_april_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
jan_april_df.index = pd.to_datetime(jan_april_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
jan_april_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
jan_april_fare_by_week = jan_april_df.resample("W").sum()
jan_april_fare_by_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# Import the style from Matplotlib.
import matplotlib.dates as mdates
from matplotlib import style
# https://www.earthdatascience.org/courses/scientists-guide-to-plotting-data-in-python/plot-with-matplotlib/introduction-to-matplotlib-plots/plot-time-series-data-in-python/
# https://matplotlib.org/stable/api/dates_api.html#matplotlib.dates.DateFormatter
from matplotlib.dates import DateFormatter
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
# Define the date format
# https://docs.python.org/3/library/datetime.html#strftime-strptime-behavior
# https://docs.python.org/3/library/datetime.html#datetime.datetime.strftime
date_form = DateFormatter("%b")
```

```python
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(25,8))
ax.plot(jan_april_fare_by_week,label=["Rural","Suburban","Urban"])
ax.set_title("Total Fare by City Type")
ax.set_ylabel("Fare($USD)")
ax.set_xlim()
ax.legend(fontsize="12", loc="center", title="type")
ax.xaxis.set_major_formatter(date_form)
# Ensure ticks fall every month 
ax.xaxis.set_major_locator(mdates.DayLocator(bymonthday=[6]))
plt.savefig("analysis/PyBer_fare_summary.png")
plt.show()
```
